```python
%run data_tools.ipynb
year = 2020
data = load_data(year)
```

```python
tpdata = data['team_pitcher'].drop(['G_ID', 'VS_T_ID', 'TB_SC', 'P_WHIP_RT'], axis = 1)
tpdata["WLS"] = tpdata["WLS"].astype("category")
```

```python
out = pd.DataFrame(data=None, columns = ['GDAY_DS', 'T_ID', 'HEADER_NO', 'GCNT', 'W', 'L', 'D', 'CG_CK', 'HOLD',
       'INN2', 'BF', 'PA', 'AB', 'HIT', 'H2', 'H3', 'HR', 'SB', 'CS', 'SH',
       'SF', 'BB', 'IB', 'HP', 'KK', 'GD', 'WP', 'BK', 'ERR', 'R', 'ER'])

for teamid in data['team']['T_ID']:
    teamdata = tpdata[tpdata['T_ID'] == teamid].reset_index().drop(['index'], axis = 1)
    teamdata = teamdata.sort_values(by = ['T_ID', 'GDAY_DS', 'HEADER_NO'], ascending = True)
    teamdata = pd.get_dummies(teamdata, columns = ['WLS'], prefix = '', prefix_sep = '', drop_first=False)
    csum_teamdata = teamdata.cumsum().reset_index()
    csum_teamdata['index'] = csum_teamdata['index'] + 1
    csum_teamdata = csum_teamdata.rename(columns = {'index': 'GCNT'})
    result = pd.concat([teamdata.filter(['GDAY_DS', 'T_ID', 'HEADER_NO']),
                       csum_teamdata.filter(['GCNT', 'W', 'L', 'D', 'CG_CK', 'HOLD', 
                       'INN2', 'BF', 'PA', 'AB', 'HIT', 'H2', 'H3', 'HR', 
                       'SB', 'CS', 'SH', 'SF', 'BB', 'IB', 'HP', 'KK', 
                       'GD', 'WP', 'BK', 'ERR', 'R', 'ER'])], axis = 1)
    out = pd.concat([out, result], sort = False)
out
```

```python
save_data(out, year, '41.팀투수(요약)')
```
